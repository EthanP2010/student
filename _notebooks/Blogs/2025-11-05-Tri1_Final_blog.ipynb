# Overview of the three sprints

### Sprint 1: Tools

This sprint was, without a doubt, the most **difficult** for me. I was frustrated as many of my tools were not working. Eventually, through **patience and perseverance**, I completed the main objective of the sprint and learned how to install and use my tools. My knowledge of using these tools increased throughout this trimester, through **new experiences** that helped strengthen my knowledge.

**These steps will forever be cemented in my brain when cloning a repo for VSCode:**


  `cd`
  
  `cd EthanP2010` --> My directory

  `git clone[link]` --> link of repo from github

  `./scripts/venv.sh` --> Activate the virtual environment

  `source venv/bin/activate` --> See venv in path

  `bundle install` --> Ensure Ruby gems for GitHub Pages is installed in (venv)

  `code .` --> Open the project in VS Code

 **3 Key things:**
 1. Setting up a new repo in VSCode
 2. Never paste in multiple commands at once (Go in order)
 3. Testing if my changes worked using `make` and commiting them in VSCode
